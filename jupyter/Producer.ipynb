# Producer | Kafka

Este notebook tem por objetivo ler o arquivo do ano mencionado e enviar os dados para o Kafka, fazendo assim a função de produtor de mensagens de streaming.

```python
# !pip install confluent-kafka
```

```python
from confluent_kafka import Producer
from confluent_kafka.admin import AdminClient
```

```python
import time
```

```python
# Define o ano do arquivo que vamos ler
ano='2022'
```

```python
#função de callback do producer. para ver se teve erro ou sucesso.

def delivery_report(err, msg):
    if err is not None:
        print('Falha : {}'.format(err))
    else:
        print('Sucesso {} [{}]'.format(msg.topic(), msg.partition()))
```

```python
# kafka configuration
conf = {
    'bootstrap.servers': 'kafka1:9092'
}
```

```python
#Remove o tópico antes de iniciar o envio
admin_client = AdminClient(conf)
admin_client.delete_topics(topics=[f"f1-{ano}"])
```

```python
# kafka producer init
producer = Producer(conf)
```

```python
# Lê o arquivo e envia para o Kafka | producer
with open(f"./dados/{ano}.csv", "r") as file:
    for line in file:
        # write messages to the users topic
        producer.produce(f"f1-{ano}", key=None, value=line.encode('utf-8'), callback=delivery_report)
        print(line)
        # time.sleep(3)
```

```python
# Espera todas as mensagens enfileiradas serem enviadas ao Kafka Producer e imprime o relatorio de callback
producer.flush()
```
