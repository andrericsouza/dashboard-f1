```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

## RACES

```python
races = pd.read_csv("./dados/races.csv")
```

```python
races.head()
```

```python
races.replace('\\N','', inplace=True)
races.head()
```

```python
races.dtypes
```

```python
races.describe(include='all').T
```

```python
# inserindo dados no postgres
engine = create_engine('postgresql://admin:admin@postgres:5432/f1') 
races.to_sql(name='dim_races', con=engine, schema='f1_schema', if_exists='replace',index=False)
```

## DRIVERS

```python
drivers = pd.read_csv("./dados/drivers.csv")
```

```python
drivers.head()
```

```python
drivers.replace('\\N','', inplace=True)
drivers.head()
```

```python
drivers.dtypes
```

```python
drivers.describe(include='all').T
```

```python
drivers[['number']] = drivers[['number']].apply(pd.to_numeric)
drivers.dtypes
```

```python
drivers.describe(include='all').T
```

```python
drivers.isna().sum()
```

```python
drivers.head()
```

```python
drivers.fillna(0, inplace=True)
drivers.head()
```

```python
drivers.isna().sum()
```

```python
# inserindo dados no postgres
engine = create_engine('postgresql://admin:admin@postgres:5432/f1') 
drivers.to_sql(name='dim_drivers', con=engine, schema='f1_schema', if_exists='replace',index=False)
```
